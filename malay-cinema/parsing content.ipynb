```python
def cast_split(cast):
    cs = cast.split(', ')
    print(cs)
    cast_1 = np.nan
    cast_2 = np.nan
    cast_3 = np.nan
    cast_4 = np.nan
    cast_5 = np.nan
    cast_6 = np.nan

    if len(cs) > 5:
        cast_1 = cs[0]
        cast_2 = cs[1]
        cast_3 = cs[2]
        cast_4 = cs[3]
        cast_5 = cs[4]
        cast_6 = cs[5]
        
    if len(cs) == 5:
        cast_1 = cs[0]
        cast_2 = cs[1]
        cast_3 = cs[2]
        cast_4 = cs[3]
        cast_5 = cs[4]
    
    elif len(cs) == 4:
        cast_1 = cs[0]
        cast_2 = cs[1]
        cast_3 = cs[2]
        cast_4 = cs[3]
        
    
    elif len(cs) == 3:
        cast_1 = cs[0]
        cast_2 = cs[1]
        cast_3 = cs[2]
        
    elif len(cs) == 2:
        cast_1 = cs[0]
        cast_2 = cs[1]
        
    elif len(cs) == 1:
        cast_1 = cs[0]


    print('cast1',cast_1,'cast2',cast_2,'cast3',cast_3,'cast4',cast_4,'cast5',cast_5, cast_6)
    
    return pd.Series({'cast_1':cast_1, 'cast_2':cast_2,'cast_3':cast_3,'cast_4':cast_4,'cast_5':cast_5, 'cast_6':cast_6})
```

```python
cast_columns = ['cast_1','cast_2','cast_3','cast_4','cast_5', 'cast_6']
df[cast_columns] = df['cast'].apply(lambda x: cast_split(x) if (type(x) == str) else pd.Series({'cast_1':np.nan, 'cast_2':np.nan, 'cast_3':np.nan, 'cast_4':np.nan, 'cast_5':np.nan}))

```

```python
def genderize(actor):
    actor_details = pd.read_csv('actor_details.csv')
    print(actor)
    if actor_details[actor_details['name']== actor].size > 0:
        if actor_details[actor_details['name']== actor]['gender'].iloc[0] == 'f':
            gender = 'Female'
        else:
            gender = 'Male'
    else:
        gender = np.nan
    return gender
```
