```python
import requests # for web requests
from bs4 import BeautifulSoup # a powerful HTML parser
from selenium.webdriver import Chrome
import pandas as pd # for .csv file read and write
import time 
import numpy as np
import re # for regular regression handling
#from requests_html import HTMLSession
#session = HTMLSession()

headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36'}
path = "/usr/local/Caskroom/chromedriver/86.0.4240.22/chromedriver"
driver = Chrome(executable_path=path)
```

```python
#key = 'site:cinema.com.my/movies/details+malay'
base_url = 'https://www.propsocial.my/buy/malaysia-1'
params = {'page': None}

links = list()
titles = list()
details = list()
prices = list()
loaded = True
page = 7100

anchors = []
while loaded:
    print('Loading page {} ...'.format(page))
    params['page'] = page
    
    driver.get(base_url+'?page='+str(page))
    #time.sleep(3)
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    search_results = soup.find_all('div', {'class': 'classified-ad-details'})
    len(search_results)
    for search_detail in search_results:
        data = search_detail.find_all('div',{'class':'classified-ad-specific-detail'})
        clean_data = list()
        for datum in data:
            clean_data.append(re.sub('\s+',' ',datum.text))
        anchors = search_detail.find_all('a')
        if len(anchors) > 0:
            link = anchors[0]['href']
            title = re.sub('\s+',' ',anchors[0].text)
        else:
            link = np.nan
            title = np.nan
        if len(anchors) > 1:
            price = re.sub('\s+',' ',anchors[1].text)
        else:
            price = np.nan
        
        links.append(link)
        titles.append(title)
        prices.append(price)
        details.append(clean_data)


   
    results = []
            
    if not len(search_results):
        loaded = False
    else:
        #position_links += links
        page += 1
```

```python
dict = {'link':links, 'title':titles, 'detail':details, 'price':prices}
df = pd.DataFrame(dict)
df
```

```python
df.to_csv('6863to7100prices.csv')
```
