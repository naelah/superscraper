```python
import requests # for web requests
from bs4 import BeautifulSoup # a powerful HTML parser
from selenium.webdriver import Chrome
import pandas as pd # for .csv file read and write
import time 
import numpy as np
import re # for regular regression handling
#from requests_html import HTMLSession
#session = HTMLSession()

headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36'}
path = "/usr/local/Caskroom/chromedriver/86.0.4240.22/chromedriver"
driver = Chrome(executable_path=path)
```

```python
df = pd.read_csv('malay_cinema_1.csv')
df
```

```python
def scrape_details(link):
#link = 'http://www.cinema.com.my/movies/details.aspx?search=2018.13349.qhaliq.26670'

    driver.get(link)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    #language = np.nan
    

    box = soup.find_all('div', {'class' : 'con-lg'})
    if box:
        content = box[0].text

        summary_1 = content.partition('Language: ')
        summary_2 = summary_1[2].partition('Subtitle: ')
        summary_3 = summary_2[2].partition('Classification: ')
        summary_4 = summary_3[2].partition('Release Date: ')
        summary_5 = summary_4[2].partition('Genre: ')
        summary_6 = summary_5[2].partition('Running Time: ')
        summary_7 = summary_6[2].partition('Distributor: ')
        summary_8 = summary_7[2].partition('Cast: ')
        summary_9 = summary_8[2].partition('Director: ')
        summary_10 = summary_9[2].partition('Format: ')
        summary_11 = summary_10[2].partition('\n')

        content = summary_1[0]
        language = summary_2[0]
        subtitle = summary_3[0]
        classification = summary_4[0]
        release_date = summary_5[0]
        genre = summary_6[0]
        running_time = summary_7[0]
        distributor = summary_8[0]
        cast = summary_9[0]
        director = summary_10[0]
        movie_format = summary_11[0]
    else:
        content = np.nan
        language = np.nan
        subtitle = np.nan
        classification = np.nan
        release_date = np.nan
        genre = np.nan
        running_time = np.nan
        distributor = np.nan
        cast = np.nan
        director = np.nan
        movie_format = np.nan

    return pd.Series({'content':content,'language': language, 'subtitle':subtitle, 'classification':classification,'release_date': release_date, 'genre':genre,'running_time':running_time,'distributor':distributor,'cast':cast,'director':director,'movie_format':movie_format})
```

```python
columns = ['content','language','subtitle','classification','release_date','genre','running_time','distributor','cast','director','format']
df[columns] = df['links'].apply(lambda x: scrape_details(x))
```
