```python
def split_add(loc):
    print(loc)
    zipc = np.nan
    state = np.nan
    country = np.nan
    states = ['Kuala Lumpur','Putrajaya','Labuan','Johor', 'Kedah', 'Kelantan', 'Pahang', 'Perak', 'Selangor', 'Terengganu', 'Perlis', 'Negeri Sembilan', 'Melaka', 'Penang','Pulau Pinang', 'Sarawak', 'Sabah']
    loc = loc.split(', ')
    for l in loc:
        if l.isnumeric():
            zipc = l
        if l in states:
            state = l
        if l == 'Malaysia':
            country = l
            
    return pd.Series({'zipcode':zipc, 'state':state, 'country':country})
```

```python
df[['zipcode','state','country']] = df.location.apply(lambda x: split_add(x) if type(x) == str else pd.Series({'zipcode':np.nan, 'state':np.nan,'country':np.nan}))
```
