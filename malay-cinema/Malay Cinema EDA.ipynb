```python
def pecah_genre(genre):
    print(genre)
    genre1 = np.nan
    genre2 = np.nan
    genre3 = np.nan
    
    if '/' in genre:
        genre = genre.split('/')
        genre1 = re.sub(r'\W+', '', genre[0].upper())
        genre2 = re.sub(r'\W+', '', genre[1].upper())
        if len(genre) >=3:
            genre3 = re.sub(r'\W+', '', genre[2].upper())
    else:
        genre1 = genre.upper()
    return pd.Series({'genre1':genre1, 'genre2':genre2, 'genre3':genre3})
    
```

```python
def romanticize(genre1, genre2):
    if ((genre1 == 'COMEDY') & (genre2 == 'ROMANCE')) | ((genre1 == 'ROMANCE') & (genre2 == 'COMEDY')):
        genre1 = 'ROMANTIC COMEDY'
    elif ((genre1 == 'COMEDY') & (genre2 == 'HORROR')) | ((genre1 == 'HORROR') & (genre2 == 'COMEDY')):
        genre1 = 'HORROR COMEDY'
    elif ((genre1 == 'COMEDY') & (genre2 == 'ACTION')) | ((genre1 == 'ACTION') & (genre2 == 'COMEDY')):
        genre1 = 'ACTION COMEDY'
    elif (genre1 == 'HISTORY')  | (genre1 == 'BIOGRAPHY') | (genre1 == 'SPORT') | (genre1 == 'PERIOD DRAMA') | (genre1 == 'PATRIOTISM'):
        genre1 = 'HISTORICAL'
    elif (genre1 == 'SCIENCEFICTION')  | (genre1 == 'SCIENCE FICTION') | (genre1 == 'FICTION') | (genre1 == 'FANTASY') | (genre2 == 'SCIENCE FICTION') | (genre2 == 'SCIENCEFICTION') :
        genre1 = 'SCIFI/FANTASY'
    elif (genre1 == 'DANCE') :
        genre1 = 'MUSICAL'
    elif (genre1 == 'MILITARY') :
        genre1 = 'ADVENTURE'
    elif (genre1 == 'EXPERIMENTAL SHORT') :
        genre1 = 'DRAMA'
    elif (genre1 == 'MYSTERY') | (genre1 == 'CRIME') | (genre2 == 'CRIME'):
        genre1 = 'THRILLER'
    elif (genre2 == 'ANIMATION'):
        genre1 = 'ANIMATION'
    elif ((genre1 == 'DRAMA') & (genre2 == 'ROMANCE')) | ((genre1 == 'ROMANCE') & (genre2 == 'DRAMA')):
        genre1 = 'ROMANCE'
    return genre1
```

```python
df[['genre1','genre2','genre3']] = df['genre'].apply(lambda x: pecah_genre(x) if type(x) == str else pd.Series({'genre1':np.nan, 'genre2':np.nan, 'genre3':np.nan}))
```

```python
def genderize_cast(cast, actors):
    print(cast)
    cast = re.sub(r'\W+', '', cast.upper())
    actors['regex'] = actors['name'].apply(lambda x: re.sub(r'\W+', '', x.upper()))
    #df['team'] =  df['team'].apply(lambda x: re.sub(r'[\n\r]*','', str(x)))
    #actors['regex'] = actors['regex'].upper()
    if type(cast) == str:
        i = actors.index[actors['regex'] == cast].tolist()
        if i: 
            #gender = actors.iloc[i]['gender']
            gender = actors['gender'].values[i][0]
            #gender = 'M'
        else:
            gender = np.nan
    else:
            gender = np.nan
    
    return gender
```

```python
def percentage_gender(df):

    gender_list = list()
    if pd.notnull(df['cast1_gender']): gender_list.append(df['cast1_gender'])
    if pd.notnull(df['cast2_gender']): gender_list.append(df['cast2_gender'])
    if pd.notnull(df['cast3_gender']): gender_list.append(df['cast3_gender'])
    if pd.notnull(df['cast4_gender']): gender_list.append(df['cast4_gender'])
    if pd.notnull(df['cast5_gender']): gender_list.append(df['cast5_gender'])
    if pd.notnull(df['cast6_gender']): gender_list.append(df['cast6_gender'])
    if pd.notnull(df['cast7_gender']): gender_list.append(df['cast7_gender'])
    if pd.notnull(df['cast8_gender']): gender_list.append(df['cast8_gender'])
    if pd.notnull(df['cast9_gender']): gender_list.append(df['cast9_gender'])
    if pd.notnull(df['cast10_gender']): gender_list.append(df['cast10_gender'])
        
    # less than 3 <3
    
    if len(gender_list) <3:
        female_dist3 = np.nan
        male_dist3  = np.nan
        female_dist4 = np.nan
        male_dist4  = np.nan
        female_dist5 = np.nan
        male_dist5  = np.nan
        female_dist6 = np.nan
        male_dist6  = np.nan
        female_dist7 = np.nan
        male_dist7  = np.nan
        female_dist8 = np.nan
        male_dist8  = np.nan
        female_dist9 = np.nan
        male_dist9  = np.nan
        female_dist10 = np.nan
        male_dist10  = np.nan
    else:
            
        # max 3

        glist = gender_list[:3]
        fem_count = glist.count('F')
        male_count = glist.count('M')
        if fem_count > 0:
            female_dist3 = fem_count / len(glist)
            male_dist3 = 1 - female_dist3
        else:
            male_dist3 = male_count / len(glist)
            female_dist3 = 1 - male_dist3    
        if (male_count + fem_count != len(glist)): print('logical error',df['title'])

        # max 4

        glist = gender_list[:4]
        fem_count = glist.count('F')
        male_count = glist.count('M')
        if fem_count > 0:
            female_dist4 = fem_count / len(glist)
            male_dist4 = 1 - female_dist4
        else:
            male_dist4 = male_count / len(glist)
            female_dist4 = 1 - male_dist4
        if (male_count + fem_count != len(glist)): print('logical error',df['title'])

        # max 5
        glist = gender_list[:5]
        fem_count = glist.count('F')
        male_count = glist.count('M')
        if fem_count > 0:
            female_dist5 = fem_count / len(glist)
            male_dist5 = 1 - female_dist5
        else:
            male_dist5 = male_count / len(glist)
            female_dist5 = 1 - male_dist5
        if (male_count + fem_count != len(glist)): print('logical error',df['title'])

        # max 6
        glist = gender_list[:6]
        fem_count = glist.count('F')
        male_count = glist.count('M')
        if fem_count > 0:
            female_dist6 = fem_count / len(glist)
            male_dist6 = 1 - female_dist6
        else:
            male_dist6 = male_count / len(glist)
            female_dist6 = 1 - male_dist6
        if (male_count + fem_count != len(glist)): print('logical error',df['title'])

        # max 7
        glist = gender_list[:7]
        fem_count = glist.count('F')
        male_count = glist.count('M')
        if fem_count > 0:
            female_dist7 = fem_count / len(glist)
            male_dist7 = 1 - female_dist7
        else:
            male_dist7 = male_count / len(glist)
            female_dist7 = 1 - male_dist7
        if (male_count + fem_count != len(glist)): print('logical error',df['title'])

        # max 8
        glist = gender_list[:8]
        fem_count = glist.count('F')
        male_count = glist.count('M')
        if fem_count > 0:
            female_dist8 = fem_count / len(glist)
            male_dist8 = 1 - female_dist8
        else:
            male_dist8 = male_count / len(glist)
            female_dist8 = 1 - male_dist8
        if (male_count + fem_count != len(glist)): print('logical error',df['title'])

        # max 9
        glist = gender_list[:9]
        fem_count = glist.count('F')
        male_count = glist.count('M')
        if fem_count > 0:
            female_dist9 = fem_count / len(glist)
            male_dist9 = 1 - female_dist9
        else:
            male_dist9 = male_count / len(glist)
            female_dist9 = 1 - male_dist9
        if (male_count + fem_count != len(glist)): print('logical error',df['title'])

        # max 10
        glist = gender_list[:10]
        fem_count = glist.count('F')
        male_count = glist.count('M')
        if fem_count > 0:
            female_dist10 = fem_count / len(glist)
            male_dist10 = 1 - female_dist10
        else:
            male_dist10 = male_count / len(glist)
            female_dist10 = 1 - male_dist10
        if (male_count + fem_count != len(glist)): print('logical error',df['title'])

    return pd.Series({'male_dist3':male_dist3, 'female_dist3':female_dist3, 'male_dist4':male_dist4, 'female_dist4':female_dist4, 'male_dist5': male_dist5, 'female_dist5': female_dist5, 'male_dist6': male_dist6, 'female_dist6': female_dist6, 'male_dist7':male_dist7, 'female_dist7':female_dist7, 'male_dist8':male_dist8, 'female_dist8':female_dist8, 'male_dist9':male_dist9, 'female_dist9':female_dist9, 'male_dist10': male_dist10, 'female_dist10': female_dist10 })
```
