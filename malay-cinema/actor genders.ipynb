```python
import pandas as pd
import numpy as np
import re
import unicodedata
```

```python
df = pd.read_csv('wiki_movies.csv')
df
```

```python
def cast_aggregate(cast_array, actor_list):
    cast = cast_array.split(', ')
    for c in cast:
        actor_list.append(c)
        print(c)
    return actor_list
```

```python
actor_list = list()
actor_list = df['cast'].apply(lambda x: cast_aggregate(x, actor_list) if type(x) == str else np.nan)
actor_list
```

```python
len(actor_list[0])
```

```python
actor_list = actor_list[0]
```

```python
actor_list
```

```python
actor_list_new = list()
for actor in actor_list:
    actor = unicodedata.normalize("NFKD", actor)
    if ',' in actor:
        actors = actor.split(', ')
        for act in actors:
            actor_list_new.append(act)
    else:
        actor_list_new.append(actor)
        
```

```python
actor_list_new
```

```python
act_df = pd.DataFrame(actor_list_new)
act_df
```

```python
act_df.columns = ['name']
```

```python
act_df.drop_duplicates(inplace=True)
```

```python
act_df.sort_values(by=['name'], inplace=True)
act_df
```

```python
act_df.to_csv('actor_details.csv')
act_df
```
